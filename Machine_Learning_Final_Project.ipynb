```python
## ONe hot encoding ##
from sklearn.preprocessing import OneHotEncoder

def perform_one_hot_encoding(df, column_name):
    dummies = pd.get_dummies(df[column_name], prefix=column_name)
    df = pd.concat([df.drop(column_name, axis=1), dummies], axis=1)
    return df

diabetes = perform_one_hot_encoding(diabetes, 'gender')
diabetes = perform_one_hot_encoding(diabetes, 'smoking_history')
```

```python
new_order = ['gender_Female', 'gender_Male', 'age', 'hypertension', 'heart_disease', 'smoking_history_No Info', 'smoking_history_current', 'smoking_history_ever', 'smoking_history_former', 'smoking_history_never', 'smoking_history_not current', 'bmi', 'HbA1c_level', 'blood_glucose_level', 'diabetes']
diabetes = diabetes[new_order]

diabetes
```

```python
## Importing packages for ML models ##
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import GridSearchCV
```

```python
## Preparing Dataset for Modeling ##
from sklearn.model_selection import train_test_split
X = diabetes.iloc[:, :13].values
Y = diabetes.iloc[:, 14]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, stratify = Y, random_state = 3 )
```

```python
## Handling class imbalance with SMOTE ##
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state = 3)
X_train_resample, Y_train_resample = smote.fit_resample(X_train, Y_train)
```

```python
## Randomized search for best parameters for XGBoost ##
from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBClassifier
import xgboost as xgb

xgb_params = {
    'learning_rate': [.001, .01, .1, .3],
    'n_estimators': [50, 100, 150, 200],
    'max_depth': [5, 10, 15, 20],
    'colsample_bytree': [0.5, 0.6, 0.7, 0.8]
}

xgb_random = RandomizedSearchCV(XGBClassifier(), param_distributions = xgb_params, n_iter = 10, cv = 5, random_state = 2)
xgb_random.fit(X_train_resample, Y_train_resample)

print('optimal parameters:', xgb_random.best_params_)
```

```python
## XGBoost Model ##
from sklearn.metrics import confusion_matrix, classification_report

xgb = xgb.XGBClassifier(n_estimators = 50, max_depth = 20, learning_rate = 0.3, colsample_bytree = 0.7, random_state = 2)
xgb.fit(X_train_resample, Y_train_resample)

print('Accuracy on training set: {:.3f}'.format(xgb.score(X_train_resample, Y_train_resample)))
print('Test score predictions:', xgb.predict(X_test))
print('Accuracy on test set: {:.3f}'.format(xgb.score(X_test, Y_test)))

## Confusion Matrix ##
Y_pred = xgb.predict(X_test)
Matrix = confusion_matrix(Y_test, Y_pred)
print(Matrix)
sns.heatmap(Matrix, annot = True, fmt = 'd')

## Plot Confusion Matrix ##
plt.title('XBGboost Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')

## Classification Report ##
print('Classification Report:\n', classification_report(Y_test, Y_pred))
```

```python
## Randomized Search for best parameters for Random Forest ##
from sklearn.ensemble import RandomForestClassifier

forest_params = {
    'n_estimators': [5, 10, 15, 20],
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 3, 5]
}

forest_random = RandomizedSearchCV(RandomForestClassifier(), param_distributions = forest_params, n_iter = 10, cv = 5, random_state = 2)
forest_random.fit(X_train_resample, Y_train_resample)

print('Optimal parameters:', forest_random.best_params_)
```

```python
## Random Forest Model ##

# Random Forest Classifier #
Forest = RandomForestClassifier(n_estimators = 10, max_depth = None, min_samples_split = 5, min_samples_leaf = 1, random_state=2)
Forest.fit(X_train_resample, Y_train_resample)

## Test Results ##

print('Accuracy on training set: {:.3f}'.format(Forest.score(X_train_resample, Y_train_resample)))
print('Test score predictions:', Forest.predict(X_test))
print('Accuracy on test set: {:.3f}'.format(Forest.score(X_test, Y_test)))

## Confusion Matrix ##
Y_pred = Forest.predict(X_test)
Matrix = confusion_matrix(Y_test, Y_pred)
print(Matrix)

## plot confusion matrix ##
sns.heatmap(Matrix, annot = True, fmt = 'd' )
plt.title('Random Forest Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')

## Classification Report ##
print('Classification Report:\n', classification_report(Y_test, Y_pred))
```

```python
## Optimal parameters for logistic regression ##
from sklearn.linear_model import LogisticRegression

Log_params = {
    'max_iter': [100, 500, 1000, 2000],
    'C': [0.001, 0.01, 0.1, 1, 10]
}

Log_grid = GridSearchCV(LogisticRegression(), param_grid = Log_params, cv = 5)
Log_grid.fit(X_train_resample, Y_train_resample)

print('optimal parameters:', Log_grid.best_params_)
```

```python
## Logistic Regression Model ##
Log = LogisticRegression(C = 0.001, max_iter = 500, random_state=3)
Log.fit(X_train_resample, Y_train_resample)

print('Accuracy on training set: {:.3f}'.format(Log.score(X_train_resample, Y_train_resample)))
print('Test set predictions:', Log.predict(X_test))
print('Accuracy: {:.2f}'.format(Log.score(X_test, Y_test)))

## Confusion Matrix ##
Y_pred = Log.predict(X_test)
Matrix = confusion_matrix(Y_test, Y_pred)
sns.heatmap(Matrix, annot = True, fmt = 'd')

## Plotting confusion matrix ##
plt.title('Logisitc Regression Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')


## Classification Report ##
print('Classification Report:\n', classification_report(Y_test, Y_pred))
```

```python
## Support Vector Machine Model ##
from sklearn.svm import SVC

svm = SVC(kernel = 'rbf', C = 10, gamma = 0.1).fit(X_train_resample, Y_train_resample)

print('Accuracy on training set: {:.3f}'.format(svm.score(X_train_resample, Y_train_resample)))
print("Test set predictions:", svm.predict(X_test))
print("Accuracy on test set: {:.2f}".format(svm.score(X_test, Y_test)))

## Confusion Matrix ##
Y_pred = svm.predict(X_test)
Matrix = confusion_matrix(Y_test, Y_pred)
sns.heatmap(Matrix, annot = True, fmt = 'd')

## Plotting confusion matrix ##
plt.title('Kernelized SVMs Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')

## Classification Report ##
print('Classification Report:\n', classification_report(Y_test, Y_pred))
```
